```python
import os
data_path = 'data'
file_paths = list()
skip_files = ['stopwords']
for root, dirs, files in os.walk(data_path):
    for file in files:
        if file not in skip_files:
            file_paths.append(f'{data_path}/{file}')
            
print(file_paths)
```

```python
novels = list[tuple]()
for file_path in file_paths:
    with open(file_path, 'r', encoding='gbk') as _f:
        tmp_novel = _f.read()
        novels.append((file_path, tmp_novel))

with open('data/stopwords', encoding='utf-8') as _f:
    stopwords = _f.readlines()

print(str(novels[0])[:512], '...')
print(stopwords[:128])
```

```python
import jieba

seg_lists = list()

for novel in novels:
    seg_list = jieba.cut(novel[1], cut_all=False)
    seg_lists.append((novel[0], [w for w in seg_list if w not in stopwords and len(w) > 1]))
    

print(seg_lists[0])
```

```python
documents = [x[1] for x in seg_lists]
```

```python
from gensim import corpora

dictionary = corpora.Dictionary(documents)
corpus = [dictionary.doc2bow(x) for x in documents]

print(corpus)
```

```python
from gensim.models import LdaModel
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

num_topics = 2

lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=num_topics,
    passes=15,
    random_state=42
)

pyLDAvis.display(gensimvis.prepare(lda_model, corpus, dictionary))
```
